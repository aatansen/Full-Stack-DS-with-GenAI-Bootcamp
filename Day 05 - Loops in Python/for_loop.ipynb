## For Loop

```python
list(range(1,11))
```

```python
for i in range(1,11):
    print(i)
```

### Step by step for loop

```python
for i in range(1,11,2):
    print(i)
```

### Reverse for loop

```python
for i in range(11,-1,-1):
    print(i)
```

```python
for i in (1,2,3,4):
    print(i)
```

### Nested for loop

```python
for i in range(1,5):
    for j in range(1,5):
        print(1,j)
```

```python
# triangle pattern using nested for loop
for i in range(1,6):
    for j in range(1,i+1):
        print("*", end="")
    print("")
```

### Loop control statements

- break
- continue
- pass

```python
# break
for i in range(1,11):
    if i==5:
        break
    print(i)
```

```python
# break
names = ["Tansen","Shakil","Bappy","Ridoy"]

for i in names:
    if i=="Tansen":
        print(i)
        break
```

```python
# continue
for i in range(1,10):
    if i==5:
        continue
    print(i)
```

```python
# pass
for i in range(1,11):
    pass
```
