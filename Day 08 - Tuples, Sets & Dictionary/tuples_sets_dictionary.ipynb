## Tuples

A tuple in Python is similar to a list. The difference between the two is that we cannot change the elements of a tuple once it is assigned whereas we can change the elements of a list.

In short, a tuple is an immutable list. A tuple can not be changed in any way once it is created.

Characterstics

- Ordered
- Unchangeble
- Allows duplicate

## Tuple Comprehension


You might have read about list comprehension in Python. However, there is no tuple comprehension in Python. But We will discuss how we can emulate tuple comprehension in Python.

```python
# Add 1 to 10 numbers to a list

L = []

for i in range(1,11):
    L.append(i)

print(L)
```

```python
tuple((i for i in range(1,11)))
```

```python
# scalar multiplication on a vector

v = [2,3,4]
s = -3

L = []

for i in v:
    result = i * s
    L.append(result)

print(L)
```

```python
TC = tuple((i*s for i in v))
print(TC)
```

```python
v=[2,3,4]
result=tuple((i*-3 for i in v ))
result
```

# Sets

A set is an unordered collection of items. Every set element is unique (no duplicates) and must be immutable (cannot be changed).

However, a set itself is mutable. We can add or remove items from it.

Sets can also be used to perform mathematical set operations like union, intersection, symmetric difference, etc.

Characterstics:
- Unordered
- Mutable
- No Duplicates
- Can't contain mutable data types

# Creating Sets

```python
# empty
s = set()
print(s)
print(type(s))
```

```python
# 1D 
s1 = {1,2,3}
print(s1)
```

### Using zip() with Sets

```python
s1 = {1, 2, 3}
s2 = {'a', 'b', 'c'}
zipped = zip(s1, s2)
print(list(zipped))  # Example output: [(1, 'b'), (2, 'a'), (3, 'c')]
```

```python
zipped = [(1, 'a'), (2, 'b'), (3, 'c')]
s1, s2 = zip(*zipped)
print(s1)  # (1, 2, 3)
print(s2) 
```

## Dictionary

Dictionary in Python is a collection of keys values, used to store data values like a map, which, unlike other data types which hold only a single value as an element.

In some languages it is known as map or assosiative arrays.

dict = { 'name' : 'tansen' , 'age' : 22 , 'gender' : 'male' }

### Characterstics:

- Mutable
- Indexing has no meaning
- keys can't be duplicated
- keys can't be mutable items

### Create Dictionary

```python
# empty dictionary
d = {}
print(d)
print(type(d))
```

```python
# 1D dictionary
d1 = {"name": "tansen", "age":22, "Gender":"male"}
print(d1)
```

```python
# with mixed keys
d2 = {(1,2,3):33, "name": "tansen"}
print(d2)
```

```python
# with mixed keys
d2 = {[1,2,3]:33, "name": "tansen"}
print(d2)
```

```python
# 2D dictionary
d3 = {
    "name": "tansen",
    "college_name": "ABC",
    "semester": 5,
    "subjects": {"DSA":80, "Math":50, "English": 70}

}

print(d3)
```

```python
# using sequence and dict function
d4 = dict([("name", "tansen"), ("age", 22)])
print(d4)
```

```python
# duplicate keys
d5 = {"name": "tansen", "name": "Alex", "age":22, "Gender":"male"}
print(d5)
```

```python
# mutable items as keys
d6 = {[1,2,3]:33, "name": "tansen"}
print(d6)
```

### Dictionary Comprehension

```python
# print 1st 10 numbers and their squares
{i:i**2 for i in range(1,11)}
```

```python
products = {'phone':10,'laptop':0,'charger':32,'tablet':0}

for i, j in products.items():
   

    if j >0:
        print(i,j)
```

```python
# using if condition
products = {'phone':10,'laptop':0,'charger':32,'tablet':0}

{key:value for (key,value) in products.items() if value>0}
```

```python
# using zip
days = ["Sunday", "Monday","Tuesday","Wednesday","Thursday","Friday","Saturday"]
temp_C = [30.5,32.6,31.8,33.4,29.8,30.2,29.9]

{i:j for (i,j) in zip(days,temp_C)}
```
