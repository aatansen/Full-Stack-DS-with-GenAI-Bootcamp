# Measure Of Central Tendency

1. Mean
2. Median
3. Mode

```python
ages = [23,24,32,45,12,43,67,45,32,56,32,420]
```

```python
import numpy as np
```

```python
print(np.mean(ages))
```

```python
print(np.median(ages))
```

```python
import statistics
```

```python
print(statistics.mean(ages))
```

```python
print(statistics.median(ages))
```

```python
print(statistics.mode(ages))
```

```python
import seaborn as sns
```

```python
sns.boxplot(ages)
```

### 5 Number Summary

```python
q1,q3=np.percentile(ages,[25,75])
```

```python
print(q1,q3)
```

```python
IQR=q3-q1
lower_fence=q1-1.5*(IQR)
higher_fence=q3+1.5*(IQR)
print(lower_fence,higher_fence)
```

# Measure Of Dispersion

1. Variance
2. Standard Deviation

```python
ages = [23,24,32,45,12,43,67,45,32,56,32]
```

```python
statistics.variance(ages)
```

```python
statistics.stdev(ages)
```

```python
statistics.pvariance(ages)
```

```python
import math
math.sqrt(statistics.pvariance(ages))
```

```python
#### Q-Q plot
import matplotlib.pyplot as plt
import scipy.stats as stat
import pylab
def plot_data(sample):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    sns.histplot(sample,kde=True)
    plt.subplot(1,2,2)
    stat.probplot(sample,dist='norm',plot=pylab)
    plt.show()
```

```python
# create a normal distributed data
s=np.random.normal(0.5,0.2,1000)
```

```python
plot_data(s)
```

```python

plot_data(df['sepal_width'])
```
