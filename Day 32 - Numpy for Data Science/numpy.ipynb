```python
# Lists vs Arrays
heights_list = [1.75, 1.68, 1.80]
weight_list = [65, 50, 75]

# This works with NumPy:
heights_arr = np.array(heights_list)
weights_arr = np.array(weight_list)

bmi = weights_arr / (heights_arr ** 2)
print(bmi)
```
