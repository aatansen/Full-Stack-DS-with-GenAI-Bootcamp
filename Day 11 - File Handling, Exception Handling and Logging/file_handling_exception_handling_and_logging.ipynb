# File Handling in Python

##### Types of data used for I/O:
- Text - '12345' as a sequence of unicode chars
- Binary - 12345 as a sequence of bytes of its binary equivalent

##### Hence there are 2 file types to deal with
- Text files - All program files are text files
- Binary Files - Images,music,video,exe files



### How File I/O is done in most programming languages

- Open a file
- Read/Write data
- Close the file

# Writing to a file

```python
#if the file is not present

f = open("files/sample.txt", "w")
f.write("Hello Python")
f.close()
```

```python
# write multiline strings
f = open('files/sample1.txt','w')
f.write('hello world')
f.write('\nhow are you?')
f.write("\nwho are you?")
f.close()
```

```python
# case 2 - if the file is already present
f = open('files/sample1.txt','w')
f.write('Yo')
f.close()
```

```python
# introducing append mode
f = open('files/sample1.txt','a')
f.write('\nI am fine')
f.close()
```

```python
# write lines
L = ['hello\n','hi\n','how are you\n','I am fine']

f = open('files/sample.txt','w')
f.writelines(L)
f.close()
```

# reading from files

```python
f = open('files/sample.txt','r')
s = f.read()
print(s)
f.close()
```

```python
# reading upto n chars
f = open('files/sample.txt','r')
s = f.read(15)
print(s)
f.close()
```

```python
# readline() -> to read line by line
f = open('files/sample.txt','r')
print(f.readline(),end='')
print(f.readline(),end='')
f.close()
```

```python
f = open('files/sample.txt','r')

while True:
    data = f.readline()

    if data == "":
        break
    else:
        print(data, end= "")


f.close()
```

# File Handling using Context Manager (With)

### Using Context Manager (With)

- It's a good idea to close a file after usage as it will free up the resources
- If we dont close it, garbage collector would close it
- with keyword closes the file as soon as the usage is over

```python
with open("files/sample.txt", "w") as f:
    f.write("Hello there!")
```

```python
# read()
with open("files/sample.txt", "r") as f:
    s = f.read()
    print(s)
```

```python
# readline()
with open("files/sample.txt", "r") as f:
    s = f.readline()
    print(s)
```

```python
# append mode
with open("files/sample.txt", "a") as f:
    s = f.write("\nMy name is Tansen")
```

```python
# moving within a file
with open("files/sample.txt", "r") as f:
    print(f.read(10))
```

```python
with open("files/sample.txt", "r") as f:
    f.seek(5)
    print(f.read())
```

```python
with open("files/sample.txt", "r") as f:
    f.seek(5)
    print(f.tell())
```

# Pickling in Python

### Pickling
`Pickling` is the process whereby a Python object hierarchy is converted into a byte stream, and `unpickling` is the inverse operation, whereby a byte stream (from a binary file or bytes-like object) is converted back into an object hierarchy.

```python
def display_info():
    return "Hi my name is Tansen and I am a Programmer"
```

```python
d = display_info()
```

```python
# pickle dump
import pickle
with open('files/function.pkl','wb') as f:
  pickle.dump(d,f)
```

```python
# pickle load
import pickle
with open('files/function.pkl','rb') as f:
  p = pickle.load(f)
  print(p)
```

### Logging in Python

```python
import logging
import os
```

```python
LOG_DIR = "logs"
LOG_FILE_NAME = "application.log"

os.makedirs(LOG_DIR, exist_ok=True)

log_path = os.path.join(LOG_DIR, LOG_FILE_NAME)

logging.basicConfig(
    filename=log_path,
    format="[ %(asctime)s ] %(name)s - %(levelname)s - %(message)s",
    level=logging.INFO
)
```

```python
def profile(name):
    if type(name) == str:
        print(f"Welcome {name}")
    else:
        logging.info("Your input is not valid!")

```

```python
profile(1234)
```

```python
profile(34343)
```
